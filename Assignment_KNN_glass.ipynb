```python
X = df.iloc[:,:9]
X
```

```python
Y = df['Type']
Y
```

```python
#Standardization
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss_x = ss.fit_transform(X)
```

```python
ss_x = pd.DataFrame(ss_x)
ss_x
```

```python
ss_x.columns = list(X)
ss_x
```

```python
#Data Partition
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(ss_x,Y,train_size = 0.7)
```

```python
#ModelFitting
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors = 8, p = 2) # p = 2 ----Eucledian distance
KNN.fit(X_train,Y_train)                           # p = 1 ----manhatten distance
```

```python
Y_pred_train = KNN.predict(X_train)
Y_pred_test = KNN.predict(X_test)
```

```python
Y_pred_test
```

```python
from sklearn.metrics import accuracy_score
print("Training Accuracy Error: ",accuracy_score(Y_train,Y_pred_train).round(2))
print("Test Accuracy Error: ",accuracy_score(Y_test,Y_pred_test).round(2))
```
